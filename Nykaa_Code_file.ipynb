# Lips

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import os 
import requests
import numpy as np 
import pandas as pd 

name_data = []
review_data = []
price_data = []
discount_data = []
og_price_data = []
image_links = []

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"}
# Chrome options (helps avoid basic bot detection)
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),
                          options=options)

for i in range(1,101):
    print(f"scraping page {i}")

    url = f"https://www.nykaa.com/makeup/lips/c/15?page_no={i}&sort=popularity&search_redirection=True"
    driver.get(url)
    time.sleep(2)

    soup = BeautifulSoup(driver.page_source,"html.parser") 

    containers = soup.find_all("div" , class_ = "css-d5z3ro")

    if not containers:
        print("NO Products found,stopping.")
        break


    for c in containers:

        name = c.find("div" , attrs = {"class" : "css-xrzmfa"})
        name_data.append(name.text.strip() if name else np.nan)

        review = c.find("span" , attrs = {"class" : "css-1qbvrhp"})
        review_data.append(review.text.strip() if review else np.nan)

        price = c.find("span" , attrs = {"class" : "css-111z9ua"})
        price_data.append(price.text.strip() if price else np.nan)

        discount = c.find("span" , attrs = {"class" : "css-cjd9an"})
        discount_data.append(discount.text.strip() if discount else np.nan)

        og_price = c.find("span" , attrs = {"class" : "css-17x46n5"})
        og_price_data.append(og_price.text.strip() if og_price else np.nan)

        
        img = c.find("img")
        if img:
            link = img.get("src") or img.get("data-src") or img.get("data-lazy")
            if link and link.startswith("/"):
                link = "https://www.nykaa.com/" + link
            image_links.append(link if link else np.nan)
        else:
            image_links.append(np.nan)


print("Scraping completed")
```

```python
print(len(name_data))
print(len(review_data))
print(len(price_data))
print(len(discount_data))
print(len(og_price_data))
print(len(image_links))
```

```python
import pandas as pd
df = pd.DataFrame({'Product Name': name_data,
                  'Original Price': og_price_data,
                  'Offer Price': price_data,
                  'Discount': discount_data,
                  'Reviews': review_data,
                  'Images Links': image_links})
df
```

```python
df.to_csv("Lips_Products.csv",index = False)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import os 
import requests
import numpy as np 
import pandas as pd 

name_data = []
review_data = []
price_data = []
discount_data = []
og_price_data = []
image_links = []

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"}
# Chrome options (helps avoid basic bot detection)
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),
                          options=options)

for i in range(1,3): #98
    print(f"scraping page {i}")

    url = f"https://www.nykaa.com/fragrance/men/c/55?page_no={i}&sort=popularity"
    driver.get(url)
    time.sleep(2)

    soup = BeautifulSoup(driver.page_source,"html.parser") 

    containers = soup.find_all("div" , class_ = "productWrapper css-17nge1h")

    if not containers:
        print("NO Products found.")
        break


    for c in containers:

        name = c.find("div" , attrs = {"class" : "css-xrzmfa"})
        name_data.append(name.text.strip() if name else np.nan)

        review = c.find("span" , attrs = {"class" : "css-1qbvrhp"})
        review_data.append(review.text.strip() if review else np.nan)

        price = c.find("span" , attrs = {"class" : "css-111z9ua"})
        price_data.append(price.text.strip() if price else np.nan)

        discount = c.find("span" , attrs = {"class" : "css-cjd9an"})
        discount_data.append(discount.text.strip() if discount else np.nan)

        og_price = c.find("span" , attrs = {"class" : "css-17x46n5"})
        og_price_data.append(og_price.text.strip() if og_price else np.nan)

        
        img = c.find("img")
        if img:
            link = img.get("src") or img.get("data-src") or img.get("data-lazy")
            if link and link.startswith("/"):
                link = "https://www.nykaa.com/" + link
            image_links.append(link if link else np.nan)
        else:
            image_links.append(np.nan)


print("Scraping completed") 
```

```python
import pandas as pd
df = pd.DataFrame({'Product Name': name_data,
                  'Original Price': og_price_data,
                  'Offer Price': price_data,
                  'Discount': discount_data,
                  'Reviews': review_data,
                  'Images Links': image_links})
df
```
